# Plotting two-dimensional data

To round up this section of data science, I wanted to demonstrate how to use a few two-dimensional plotting methods which are useful for viewing different kinds of data.

Here, I am loading data from an Magnetic Resonance Image (MRI) scan of a healthy brain. This image was taken from a study completed to analyse brain response to music ([details here](https://openneuro.org/datasets/ds003720/versions/1.0.0), [data here](https://github.com/OpenNeuroDatasets/ds003720)) available through the [OpenNeuro project](https://openneuro.org).

```python
import numpy as np
import h5py

# This data has been stored in a HDF5 file to save space
# This code reads that file and converts this to a numpy array for us to use
data_file = "data/mri_scan1.h5"
hf = h5py.File(data_file, 'r')

table = 'mri_1'
mri = hf.get(table) # Extract table from HDF5 file
mri = np.array(mri) # Convert to a numpy array

hf.close()
```

We have read the data from this MRI scan and converted this into a numpy array. The data is three-dimensional and, as with any numpy array, we can look at its shape:

```python
mri.shape
```

### Plotting an image

We can explore this data by taking slices in each dimension and using the [`imshow()` function](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.imshow.html) provided by matplotlib. The `imshow()` function accepts two-dimensional inputs and can display these on a grid. In this case, we have also included the `origin` input and set this to "lower" to make sure (0,0) is in the bottom left corner.

```python
import matplotlib.pyplot as plt

fig, (ax_0, ax_1, ax_2) = plt.subplots(nrows=1, ncols=3, figsize=(10,6))

# Select one slice (index 100) on axis 0 and plot
mri_slice_0 = mri[100, :, :]
ax_0.imshow(mri_slice_0, origin="lower")

# Select one slice (index 100) on axis 1 and plot
mri_slice_1 = mri[:, 100, :]
ax_1.imshow(mri_slice_1, origin="lower")

# Select one slice (index 100) on axis 2 and plot
mri_slice_2 = mri[:, :, 100]
ax_2.imshow(mri_slice_2, origin="lower")
```

From looking at the images above, we can start to build a picture of the data. A slice along axis 0 is side-to-side across the head, along axis 1 is back-to-front and along axis 2 is bottom-to-top.

#### Filled contour plot

If we consider the bottom-to-top direction, which we could term the z axis, we can look at an average across this dimension. We could use imshow to do this, but as an alternative we could plot this a filled contour plot using the [`ax.contourf()` function](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.contourf.html)

This also accept two-dimensional inputs but can be more flexible that `imshow`

```python
fig, ax = plt.subplots()

# Can take a mean along a specific axis by passing an axis input
mri_avg_z = mri.mean(axis=2)
ax.contourf(mri_avg_z)
```

This is one way to show a density plot, in this case of the scan in the top-down direction.

### Animation

From this data, we can create a simple animation, moving through each slice in one direction. Animation is something that will be covered in more detail later in the course. This is a demonstration of how this could be done with our data using `matplotlib`.

If you have a look at the code below, you can see I've done this using the [`animation` sub-module](https://matplotlib.org/stable/api/animation_api.html) available through matplotlib, though there are many other options in Python and otherwise for this sort of animation.

```python
import matplotlib.pyplot as plt
from matplotlib import animation

# Define our plotting area
fig, ax = plt.subplots()

# Define data used within function.
data = mri

# Define number of frames
num_dimension = data.shape[2]
num_frames = num_dimension

# Define time for each frame in the animation
frames_per_second = 20
interval = 1000/frames_per_second # time for each frame in milliseconds

def animate(i):
    ax.clear() # Make sure axis is cleared for next plot
    
    # Extract data slice based on index and plot using imshow
    data_slice = data[:,:,i]
    img = ax.imshow(data_slice, origin="lower") # returns image attached to an axis (AxesImage object)
    
    # Can return objects which have been plotted - needed for some FuncAnimation options
    # return (img, )

# Create and save animation
anim = animation.FuncAnimation(fig, animate, frames=num_frames, interval=interval)
anim.save("mri_scan.gif")
```

***Note**: This animation may or may not run within your Jupyter notebook depending on which system (e.g. Windows/MacOS/Noteable) you are using. For all systems, you should be able to open the associated animation file "mri_scan.gif" to view this animation.*

To generate my animation, I have used a [matplotlib function called `FuncAnimation`](https://matplotlib.org/stable/api/_as_gen/matplotlib.animation.FuncAnimation.html). The arguments I have passed are:
- my plotting canvas, `fig`
- name of a function to use for each frame of the animation, `animate`
- number of frames
- delay between frames

If we consider that second input, the concept of passing the name of a function is similiar to the `apply` function we looked at for use with pandas DataFrames. Again, this links to the idea of *functional programming* where functions themselves are objects which can be passed around and used. 

In this case we have created a function called `animate`. For use with `FuncAnimation`, there are a few expectations around inputs and outputs for this function:
 - Input: should accept the frame number as an integer
 - Output: None needed in this case (but sometimes needs plotted objects returned)

The `animate` function will be called for each frame in our animation - each time being passed a number (which we have called `i`) which corresponds to the frame number. In our case we can set the total number of frames to be equal to our bottom-to-top dimension and use the frame number to slice our array and plot one 2D image at a time.

We can also use this information to define both the number of frames, `num_frames` and the delay between frames, `interval` for our animation.
