# Model validation

Last week we started talking about machine learning, using the relatively simple linear regression model as a test case. We also discussed residuals, and how they can be used to help identify trends and begin to evaluate the suitability of various modelling techniques.

So far, though, we have not been truly evaluating the performance of our model. In order to do this effectively, we need to check how our model performs when used to predict **new** data. Here _new data_ means data that the model was not trained on. This is important to avoid something called _overfitting_, where a model learns the training data so completely that it is unable to generalise to new data.

It may seem obvious, but the safest way to avoid overfitting is to split the dataset that you use to build your model into two distinct sets: a _training_ set and a _testing_ set. You then train your model on your training set, but you evaluate the performance of your model using your testing set. We'll be experimenting with this function in the following, and we'll get started using the familiar penguins dataset from last week.

Run the cell below to load the dataset.

```python
import pandas as pd
import seaborn as sns
%matplotlib inline

penguins = sns.load_dataset("penguins")
penguins = penguins.dropna()
penguins.info()
```

We're going to be setting up a linear regression model for the two most highly correlated fields in the penguins dataset: flipper length and body mass.

```python
sns.relplot(data=penguins,x="flipper_length_mm",y="body_mass_g")
```

## Train Test Split

To help us split our data in this way, scikit-learn has a helper function [`train_test_split()`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html). This function will split your data into random train and test samples for you. Because the samples are chosen randomly, this function has an argument called `random_state` which must be set in order to ensure reproducibility. If `random_state` is not set, then your training and testing datasets will contain different subsets of the input dataset each time you call `train_test_split()`.


#### Exercise 1

To demonstrate the function of the `random_state` argument, try running `train_test_split()` twice on the same input data in the cell below. You will need to:
- add a second call to `train_test_split()` in the first cell
- replace the `????` with your second test set in the second cell

Do you get the same subset of data in the testing dataset each time?

```python
# input data for experimenting with train_test_split
Y = penguins["body_mass_g"]

# import train_test_split
from sklearn.model_selection import train_test_split

# train_test_split call 1
y_train_1, y_test_1 = train_test_split(Y)

# train_test_split call 2
y_train_2, y_test_2 = train_test_split(Y)

```

```python
# check whether the two resulting testing datasets are the same
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

# plot your test sets against one another to see whether or not they match
# here we sort the arrays, since train_test_split samples Y randomly 
# (elements will not necessarily be returned in the same order)
ax.scatter(np.sort(y_test_1),np.sort(y_test_2))
```

Another important argument is the `test_size`. This determines the proportion of your input data that you would like to save for testing. If this argument is not set (as was the case in the example above) then the default is 0.25.

#### Exercise 2

**(a)** From the penguins dataset, we are going to use flipper length to predict body mass. Prepare the data for this regression model. To do this you will need to carry out the following steps:
- Create an input dataframe for flipper length and series for body mass (recall that you did this last week when we looked at regression models).
- Import `train_test_split()` from scikit-learn.
- Separate these input datasets into training and testing datasets using `train_test_split()`.
    - Use 30% of your input data for testing.
    - Be sure to specify a `random_state` for reproducibility. _(If you want to conform to tradition, this is often set to 42)._

```python
flipper_length_mm = penguins[["flipper_length_mm"]]
body_mass_g = penguins["body_mass_g"]
train_mass, test_mass = train_test_split(body_mass_g, test_size=0.3,random_state=42)
train_flipper, test_flipper = train_test_split(flipper_length_mm, test_size=0.3,random_state=42)
```

**(b)** Edit the code block below to do the following:
- train the model on the training data.
- make predictions using the testing data.

```python
# create an instance of the model
from sklearn import linear_model
model = linear_model.LinearRegression(fit_intercept=True)

# train the model using the *training* set
model.fit(train_flipper, train_mass)

# use the model to predict y-values of the *testing* set
mass_predict = model.predict(test_flipper)
```

**(c)** Use the test data and corresponding predictions to plot the residuals from this model as a function of flipper length. 

Recall from last week that to obtain the residuals we take the difference between the true (or observed) value of $y$ and the predicted value of $y$.

```python
fig, ax = plt.subplots()
ax.plot(test_flipper, test_mass - mass_predict,".")
```

## Testing metrics

In order to quantify the performance of our models, we can make use of metrics that are built in to scikit-learn. For example, the $R^{2}$ score or [_coefficient of determination_](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html#sklearn.metrics.r2_score) is one metric used to determine how closely the model result agrees with the data. This method can be called from the `metrics` module in scikit-learn, and takes as input two arguments: 
1. The true $y$-values
2. The predicted $y$-values

A perfect model would get an $R^{2}$ score of 1.0.

It's also a good idea to look at the [_mean squared error_](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html) (MSE). The root mean squared error (RMSE) tells us the average difference between the predicted values and the actual values.  


#### Exercise 3

**(a)** Replace the `????` below to print the mean squared error and $R^{2}$ score for this model.

```python
from sklearn.metrics import r2_score, mean_squared_error
y_true = test_mass
y_pred = mass_predict
print(f"r2_score: {r2_score(y_true, y_pred)}")
print(f"root mean squared error: {mean_squared_error(y_true, y_pred)}")
```

```python
print(f"root mean squared error: {mean_squared_error(y_true, y_pred, squared=False)}")
```
