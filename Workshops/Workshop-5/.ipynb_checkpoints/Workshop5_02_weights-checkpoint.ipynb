# Working with sample weights

So far we've worked with datasets in which all observations are equally weighted. In other words, we assume that every observation or measurement is just as reliable (and as relevant) as any other in the dataset. When working with experimental data, however, this can often not be the case. 

#### Example case
For a simple example, consider the following:

We have administered a student satisfaction survey to 100 students taking chemistry, and 200 students taking physics. In physics, 150 students rated overall experience above 70%, while in chemistry 75 students made ratings above this threshold. 

Based on the absolute number of satisfied students, we might say that physics had a greater number of satisfied students than chemistry. Of course, this fails to account for the number of students in each _sample_. 

If we were to _weight_ (or _normalise_) the survey responses to account for the different number of participants from physics and chemistry, we would find that there is no significant difference between the two samples. This is because physics had twice the number of survey participants that chemistry did, and so we apply a sample weight for survey responses from physics of 0.5.

## Group Activity

In a gamma ray spectroscopy laboratory session, students are studying the attenuation of gamma radiation emitted from Caesium-137 (Cs-137) as it is shielded by a metal absorber. To do this, students use an experimental setup that records the number of gamma-ray emissions detected from the Cs-137 source over a period of time:

- For their first measurement, there is no shielding material standing between the Cs-137 source and the detector. 
- For all subsequent measurements, the students incrementally insert metal sheets of thickness 4 mm between the Cs-137 source and the detector. 

There is a schematic of the experimental setup below, with four absorber plates pictured between the radioactive source and the detector.

![Alt](img/am36.png "Schematic of experimental setup")

As the thickness of material between the Cs-137 source and the detector setup increases, the intensity of the gamma radiation drops off exponentially according to:

$I = I_0 \exp^{-\mu x}$,

where $I$ is the measured intensity, $I_0$ is the initial measured gamma intensity (the intensity of the gamma-ray source in the absence of any material shielding), $x$ is the thickness of the absorbing material and $\mu$ is the mass attenuation coefficient of the material.

This experiment is carried out once using tin as the absorbing material, and once using lead. The Cs-137 source remains the same throughout.

Run the cell below to load the experimental measurements from this laboratory.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = pd.DataFrame(
    {
    "Thickness_mm":[0,4,8,12,16,20,24,28,0,4,8,12,16,20,24,28],
    "Material":["tin","tin","tin","tin","tin","tin","tin","tin","lead","lead","lead","lead","lead","lead","lead","lead"],
    "Counts":[199431,161923,130048,104165,85486,71214,56188,46485,119659,76948,49503,32363,20402,13505,8695,6443],
    "Error":[8290,8180,8090,6730,6840,10140,5690,6330,4970,3680,3450,3410,3800,3080,2500,2950],
    "Time_s":[100,100,100,100,100,100,100,100,60,60,60,60,60,60,60,60]
    }
)
```

Below is a plot showing the drop in radiation intensity (measured as the number of gamma ray emissions counted by the experimental detector setup) as a function of the thickness of the absorbing material. There are two lines, one for tin and one for lead.

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(data[data["Material"]=="tin"]["Thickness_mm"],
        data[data["Material"]=="tin"]["Counts"], 
        label="tin")
ax.plot(data[data["Material"]=="lead"]["Thickness_mm"],
        data[data["Material"]=="lead"]["Counts"], 
        label="lead")

ax.set_xlabel("Thickness (mm)")
ax.set_ylabel("Counts")
ax.legend()
```

Since lead is more dense than tin, we would expect the intensity of gamma radiation to drop off more rapidly when exposed to lead absorbers. It is difficult to confirm this by eye from the plot above.

## Weights as uncertainties

Sample weights can be important when building models. Often models assume that you will have a data sample that is balanced. Apart from sample normalisation, though, weights can also be useful in regression models to deal with uncertainties, or variance, in your measurements. 

Take a look at the example data below, which consists of another student dataset. This time, students are working to find the spring constant of a spring, using Hooke's Law:

$F = -kx$,

where $F$ is the force exerted on the spring, $x$ is the extension to the spring as a consequence of said force, and $k$ is the spring constant. 

In the laboratory, students have been exerting increasing amounts of force to the spring and measuring the subsequent extension. They have an experimental setup that is supposed to detect and measure the spring extension without requiring them to intervene manually, but it looks as though the extension measurement isn't always reliable.

```python
data = pd.read_csv("data/noise_line.csv")

import seaborn as sns
sns.relplot(data=data,x="Force_N",y="Extension_cm")
```

```python
data["Error_cm"]
```

Here we see what appears to be an underlying linear model (which we would expect), but there is clearly some noise in the measurements. There are also a few fairly significant outliers at high extension.

If we look at the residuals plot below, we also see that the residuals seem to be dependent on the force applied. In other words, the variance in the measurements increases as the force applied to the spring increases. This is not ideal, as it means that our residuals depend on our input variable.

```python
sns.residplot(data=data,x="Force_N",y="Extension_cm")
```

The extension measurements have known uncertainties, as shown below. We can see that the large outliers do indeed have correspondingly large uncertainties associated with them.

```python
f, ax = plt.subplots(figsize=(8,8))
ax.errorbar(data["Force_N"],data["Extension_cm"],abs(data["Error_cm"]),fmt=".")
```

#### Exercise 1

First, let's try a linear regression, using force to predict the extension to the spring. Edit the code below to do this. Use the following settings for `train_test_split()`:
- Use 30% of the data for testing, and the rest for training.
- Set the `random_state` to 42.

```python
# prepare the model data
X = data[["Force_N"]]
Y = data["Extension_cm"]

# separate data into training and testing sets
from sklearn.model_selection import train_test_split
(x_train, x_test), (y_train, y_test) = train_test_split(X, test_size=0.3,random_state=42), train_test_split(Y, test_size=0.3,random_state=42)

# set up a linear regression model
from sklearn import linear_model
model = linear_model.LinearRegression(fit_intercept=True)

# fit the model on the training data
model.fit(x_train, y_train)

# test the model performance using the test data
test_predict = model.predict(x_test)

# print some performance metrics
from sklearn.metrics import r2_score, mean_squared_error
print(f"unweighted model r2_score: {r2_score(y_test, test_predict)}")
print(f"unweighted model root mean squared error: {mean_squared_error(y_test, test_predict,squared=False)}")

```

Based on the printed model metrics alone, it looks as though the above model does a reasonable job of reproducing the measured data. However, is this model the best that we can do in terms of capturing the underlying model in the data?

We can try using sample weights to down-weight measurements that we know have high uncertainties when training our model. To do this, we will weight each measurement by the inverse of the square of the measurement uncertainty:

$w = \frac{1}{\sigma^{2}}$

This method penalises outliers by giving a much lower weight than more reliable data points. As a result the model does not put as much importance on closely reproducing the data points with large measurement uncertainties as it does the more well-measured points.

#### Exercise 2
**(a)** Let's try using sample weights in our linear regression model. First, add a new column to the dataframe called "weight" and set the values of that column according to the formula noted above.

```python
mse = mean_squared_error(y_test, test_predict, squared=False)
weight = 1 / mse ** 2
data["weight"] = np.ones(data.shape[0])
data["weight"] *= weight
data
```

**(b)** Modify the code below to take into account the sample weights, and print out the performance metrics at the end.

```python
# prepare the model data
X = data[["Force_N"]]
Y = data["Extension_cm"]
W = data["weight"]

# separate data into training and testing sets
from sklearn.model_selection import train_test_split
(x_train, x_test), (y_train, y_test), (w_train, w_test) = train_test_split(X, test_size=0.3,random_state=42), train_test_split(Y, test_size=0.3,random_state=42), train_test_split(W,test_size=0.3,random_state=42)

# set up a linear regression model
from sklearn import linear_model
model_w = linear_model.LinearRegression(fit_intercept=True)

# fit the model on the training data **including weights in the training**
model_w.fit(x_train, y_train, sample_weight=w_train)

# test the model performance using the test data
test_predict = model_w.predict(x_test)

# print some performance metrics
from sklearn.metrics import r2_score, mean_squared_error
print(f"weighted model r2_score: {r2_score(y_test, test_predict)}")
print(f"weighted model root mean squared error: {mean_squared_error(y_test, test_predict, squared=False)}")
```
