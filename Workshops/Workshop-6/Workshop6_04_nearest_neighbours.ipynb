## Bonus Exercise: predicting categorical information

This exercise is for interest and illustration only, to give an example of using numerical data to predict categorical information. 

To do this we are going to use a method called _k-nearest neighbours_, or _kNN_. In this method, if we want to predict the category of a given data point, the kNN algorithm looks for the $k$ nearest neighbours of that point (where $k$ is a number, a _hyperparameter_ in fact). In other words, the algorithm searches for a pre-defined number of data points that are closest in to the data point that we are trying to categorise. The category predicted for your data point is then assigned based on the category that was found to be most frequently associated with the data points in the nearest neighbour sample.


#### Example

See the below example for Bob, a friend or ours who is trying to choose a restaurant. We can see that Bob prefers cheaper restaurants to expensive ones, but he might be willing to go more expensive so long as the restaurant is not too far from where he lives. In fact, Bob's restaurant preferences form clear clusters in the price -- location plane defined in the figure below. 

If we take a new restaurant (displayed as a green data point) that Bob has never been to before, and we want to predict whether or not Bob is likely to like that restaurant, we can look to that restaurants _nearest neighbours_ in the price -- location plane.

Assuming we set our hyperparameter $k=4$ in this example, we find four nearest neighbours in the dashed circle drawn around our green data point below. We can see that while Bob doesn't like one of those neighbouring restaurants, he _does_ like three of them. For this reason, we would think that Bob might just add this new restaurant to his list of go-to restaurants, because we predict that he **would like it**. So, in this example, the green point would become a blue one, according to our model.

#### Exercise 1

Let's go back to our pengiuns dataset, where we will experiment with using k-nearest neighbours by taking bill depth and bill length to predict the species of penguin.


```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins = penguins.dropna()
```

Below we can see the relationship between these two variables, and we can also see that the three species of penguin form fairly well-defined clusters within the bill depth versus bill length plane.

```python
%matplotlib inline
sns.relplot(data=penguins,x="bill_depth_mm",y="bill_length_mm", hue="species")
```

First, let's create a training data _DataFrame_ and target values _Series_ and split them into train and test sets (as we have done in previous model setups).

**(a)** Run the cell below to do this.

```python
X = penguins[["bill_depth_mm","bill_length_mm"]]
Y = penguins["species"]


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

Now we will instantiate our model, but this time we will not be using a linear regression, but a [`KNeighborsClassifier`](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html). 

**(b)** Look at the code block below. Notice the following:
- the steps we are taking below are analogous to the steps we have taken in the past when setting up linear regression models
- we are setting a _hyperparameter_ `n_neighbors` _(do you remember what this does?)_
- since we are predicting categories rather than numerical information, we are using a new metric, [_Accuracy_](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html) to test our model performance.

```python
# import the relevant class from sklearn
from sklearn.neighbors import KNeighborsClassifier

# instantiate the kNN model
knn = KNeighborsClassifier(n_neighbors=5)

# train our model using our training set
knn.fit(x_train, y_train)

# make predictions using our testing set
test_pred = knn.predict(x_test)

# check the performance of the model on your testing set
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, test_pred))
```

**(c)** As always, it helps to visualise what we've just done. To do so, we're first going to use the model to make predictions using the entire input training dataset. These predictions will then be entered into a new column in the penguins DataFrame. 

Run the cell below to do this:

```python
# Make predictions using the full set of input features
y_pred = knn.predict(X)

# Create a new column to hold the predictions
penguins["predictions"] = y_pred
```

We will now make two plots to show the relationship between bill depth and bill length. In the first we will plot each species in a different colour, and in the second we will plot the _predicted species_ in a different colour. 

**(d)** Run the cell below to create these plots. Compare the two. 
- Can you spot any differences?
- Where do the predicted values differ from the actual categories?
- Can you play with the hyperparameters of the model to see whether this makes any difference to the model accuracy, and is that difference subsequently reflected in the plots below?

```python
# plot the input features with different colour per true species
sns.relplot(data=penguins,x="bill_depth_mm",y="bill_length_mm", hue="species")

# plot the input features with different colour per predicted species
sns.relplot(data=penguins,x="bill_depth_mm",y="bill_length_mm", hue="predictions")
```
